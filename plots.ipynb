```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stat
```

```python
%matplotlib widget
a = 7.34
x = np.linspace(0,
                stat.gamma.ppf(0.99, a), 100)
y = stat.gamma.pdf(x, a)

fig = plt.gcf()
# fig.set_size_inches(16, 8)
# fig.set_dpi(100)
ax = plt.subplot(1,1,1)

ax.plot(x, y + 0.02,
       'k-', lw=2, label='gamma pdf')

line_y = [0, 0.18]
# x min
ax.plot([2] * 2, [0, 0.0190],
       'k--', lw=1, label='')
# x avg
ax.plot([8] * 2, line_y,
       'k--', lw=1, label='')
# l_p
ax.plot([4] * 2, [0, stat.gamma.pdf(4, a) + 0.02],
       'k--', lw=1, label='')

# filled area
fill_x = np.linspace(0, 4, 100)
fill_y = stat.gamma.pdf(fill_x, a) + 0.02

ax.fill(np.concatenate((fill_x, [4])), np.concatenate((fill_y, [0.02 - 0.001])), fill=False, hatch='///', label='filled gamma pdf')

ax.annotate(r"$F=1-\gamma$",
            xycoords='axes fraction', xy=(0.25, 0.25),
            xytext=(0.05, 0.4),
            arrowprops=dict(arrowstyle='-'))
ax.annotate(r"$R_{\gamma}=\gamma$",
            xycoords='axes fraction', xy=(0.64, 0.25),
            xytext=(0.75, 0.6),
            arrowprops=dict(arrowstyle='-'))


# annotation
v_center = (max(y) - min(y)) / 2.0
ax.annotate(r"$f(x)$", xycoords='axes fraction', xy=(0, 0), xytext=(-0.025, 1), arrowprops=dict(arrowstyle='<-'))
ax.annotate(r"$x$", xycoords='axes fraction', xy=(-0.005, 0.1), xytext=(1, 0.1-0.015), arrowprops=dict(arrowstyle='<-'))
ax.annotate(r"$l$", xycoords='axes fraction', xy=(-0.005, 0.005), xytext=(1, -0.015), arrowprops=dict(arrowstyle='<-'))

ax.set_xticks([2, 4, 8],
              [r"$l_p$", r'$l_p$', r'$\overline{l}$'])
ax.set_yticks([])
ax.set_xlim(x[0] - 0.5, x[-1] + 1)
ax.set_ylim(0)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax2 = ax.twiny()
ax2.set_yticks([])
ax2.set_xticks([4, 8],
              [r'$x_{\gamma}$', r'$\overline{x}$'])
ax2.set_xlim(x[0] - 0.5, x[-1] + 1)
ax2.set_ylim(0)    
ax2.xaxis.set_ticks_position('bottom') # set the position of the second x-axis to bottom
ax2.xaxis.set_label_position('bottom') # set the position of the second x-axis to bottom
ax2.spines['bottom'].set_position(('axes', 0.1))
plt.setp(ax2.xaxis.get_majorticklabels(), ha='left')

for s in ["top", "right", "bottom", "left"]:
    ax.spines[s].set_visible(False)
    ax2.spines[s].set_visible(False)    
```

```python
%matplotlib widget

fig = plt.gcf()
ax = plt.subplot(1,1,1)

x = np.linspace(0, 4, 100)
def y(x):
    return ( x * 0.0001 ) ** 2 + 7e-8
ax.plot(x, y(x), 'k-', lw=2)

ax.annotate(r"$y$", xycoords='axes fraction', xy=(0, 0), xytext=(-0.01, 1), arrowprops=dict(arrowstyle='<-'))
ax.annotate(r"$x$", xycoords='axes fraction', xy=(-0.005, 0.2), xytext=(1, 0.2-0.015), arrowprops=dict(arrowstyle='<-'))
ax.annotate(r"$l$", xycoords='axes fraction', xy=(-0.005, 0.005), xytext=(1, -0.015), arrowprops=dict(arrowstyle='<-'))

# dashes
ax.plot([2.5, 2.5], [0, y(2.5)], 'k--', lw=1)
ax.plot([0, 2.5], [y(2.5)] * 2, 'k--', lw=1)

ax.plot([3.8] * 2, [5e-8, y(3.8)], 'k--', lw=1)
ax.plot([0, 3.8], [y(3.8)] * 2, 'k--', lw=1)

ax.set_xticks([2.5],
              [r"$\overline{l}_0$"])
ax.set_yticks(
    [y(0), y(2.5), y(3.8)],
    [r"$y_H$", r"$y_\delta$", r"$\overline{y}_n$"])
ax.set_xlim(x[0], x[-1] + 1)
ax.set_ylim(0)

ax2 = ax.twiny()
ax2.set_yticks(
    [y(0), y(2.5), y(3.8)],
    [r"$y_H$", r"$y_\delta$", r"$\overline{y}_n$"])
ax2.set_xticks([3.8],
              [r"$\overline{x}_p$"])
ax2.set_xlim(x[0], x[-1] + 1)
ax2.set_ylim(0)
ax2.xaxis.set_ticks_position('bottom') # set the position of the second x-axis to bottom
ax2.xaxis.set_label_position('bottom') # set the position of the second x-axis to bottom
ax2.spines['bottom'].set_position(('axes', 0.2))
plt.setp(ax2.xaxis.get_majorticklabels(), ha='right')
for s in ["top", "right", "bottom", "left"]:
    ax.spines[s].set_visible(False)
    ax2.spines[s].set_visible(False)
```
